## Toy Datasets
http://scikit-learn.org/stable/datasets/index.html

Toy datasets are loaded in Bunch objects. One could make custom Bunch objects, but this is not necessary. One could load custom data from files or databases into [NumPy arrays](http://www.numpy.org/) or [Pandas data frames](http://pandas.pydata.org/) and then use the same SciKit learn tools to manipulate the vectors...

```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
# Mostrar la salida de Matplotlib en el notebook
%matplotlib inline

housing = datasets.load_boston()
#print(housing.DESCR)
X, y = housing.data, housing.target
print("Inputs:\n{}".format(X[:3]))
print("Targets:\n{}".format(y[:3]))
```

## Create Synthetic Data 

This can be helpful to test or explore Regression and Classification algorithms...

```python
fake_regression_data = datasets.make_regression(n_samples=100, n_features=10, n_informative=5, n_targets=2, bias=1.0)
X = fake_regression_data[0]
y = fake_regression_data[1]
print(X[:3])
print(y[:3])
```

```python
fake_classification_set = datasets.make_classification(weights=[0.1])
print("Weights parameter create an unbalanced dataset (~90% belong to one class):\n{}".format(np.bincount(fake_classification_set[1])))
print("First three 'objects':\n{}".format(fake_classification_set[0][:3]))
print("Classes  where the 'objects' belong:\n{}".format(fake_classification_set[1]))
```

```python
X,y = datasets.make_blobs(n_samples=200, n_features=2, centers=6)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25, edgecolor='k')
```

## Scaling and Normalizing Data

```python
from sklearn import preprocessing as pp
boston = datasets.load_boston()
X,y = boston.data, boston.target
print(X[:,:5].mean(axis=0)) # This takes the mean value of the first 5 features
print(X[:,:5].std(axis=0)) # This takes the standard deviation of sample points of the first 5 features
print("\nThe dataset can be transformed to hold mean close to ZERO and Standard Deviation of ONE")
X_scaled = pp.scale(X[:,:3])
print(X_scaled[:,:5].mean(axis=0))
print(X_scaled[:,:5].std(axis=0))
```

## Binary Features & Thresholding
One can automatically binarize the target values to a hard class tag according to a threshold of the dataset (for example assign Tag 0 to all samples whose target is below the dataset mean and Tag 1 otherwise...)

```python
binarized_y = pp.binarize([boston.target], threshold=boston.target.mean())
print(y[:5])
print(y.mean())
print(binarized_y[:,:5])
```

## Categorical Variables: One-Hot Encoding

```python
iris = datasets.load_iris()
X,y = iris.data, iris.target
dataset = np.column_stack((X,y)) #creates a single matrix with the last column being the Class Tag
print(dataset[:3])
text_encoder = pp.OneHotEncoder()
y_one_hot = text_encoder.fit_transform(dataset[:,-1:]).toarray()[:3]
```
